```python
df1
```

```python
age_groups = {
    '<18': (0, 17),
    '18-24': (18, 24),
    '25-34': (25, 34),
    '35-44': (35, 44),
    '45-54': (45, 54),
    '55-64': (55, 64),
    '65+': (65, 120)
}

def map_age_to_group(age):
    for group, age_range in age_groups.items():
        if age_range[0] <= age <= age_range[1]:
            return group
    return 'Unknown'

df1['Age_Groups'] = df1['Age'].apply(map_age_to_group)
```
